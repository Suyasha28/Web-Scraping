```python
from bs4 import BeautifulSoup
import requests

def scrape_with_beautiful_soup(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract title
    title_element = soup.find("h1", class_="entry-title")
    title = title_element.text.strip() if title_element else "No title found"

    # Extract article text
    article_text_elements = soup.find_all("div", class_="td-post-content tagdiv-type")  # Replace with the actual class name
    article_text = "\n".join([p.get_text().strip() for p in article_text_elements])

    return title, article_text

# URL of the article
url = "https://insights.blackcoffer.com/rise-of-telemedicine-and-its-impact-on-livelihood-by-2040-3-2/"

# Call the scraping function
title, article_text = scrape_with_beautiful_soup(url)

# Print the results
print("Title:", title)
print("Article Text:", article_text)
```

```python
import re

# Function to count syllables in a word
def count_syllables(word):
    vowels = "AEIOUaeiou"
    count = 0
    if word.endswith(("es", "ed")):
        word = word[:-2]
    prev_char = ""  # Initialize prev_char to an empty string
    for char in word:
        if char in vowels and prev_char not in vowels:
            count += 1
        prev_char = char
    if word.endswith("e"):
        count -= 1
    if count == 0:
        count = 1
    return count


# Count personal pronouns using regex
pronoun_regex = r'\b(I|we|my|ours|us)\b'
personal_pronouns_count = len(re.findall(pronoun_regex, cleaned_text, flags=re.IGNORECASE))

# Calculate average word length
total_chars = sum(len(word) for word in cleaned_words)
average_word_length = total_chars / total_cleaned_words

# Calculate syllable count per word
syllable_counts = [count_syllables(word) for word in cleaned_words]

# Print the calculated values
print("Syllable Counts Per Word:", syllable_counts)
print("Personal Pronouns Count:", personal_pronouns_count)
print("Average Word Length:", average_word_length)
```
